```python
from datetime import datetime
datetime.strptime(data, '%Y-%m-%d %H:%M:%S')
```

```python
df['Data']=pd.to_datetime(df['Data'])
```

```python
df.info()
```

```python
df['Tempo']=pd.to_timedelta(df['Tempo'])
```

```python
df['Tempo']=df['Tempo'].dt.total_seconds()
```

```python
df['Tempo em movimento']=pd.to_timedelta(df['Tempo em movimento']).dt.total_seconds()
```

```python
metros=[]
for i in df['Distância']:
    i=i.replace(',','')
    if i == '--':
        i=i.replace('--','0')
    if '.' in i:
        i=float(i)*1000
    else:
        i=float(i)
    metros.append(i)
df['Distância']=metros
```

```python
df=df.loc[df['Tempo em movimento']!=0]
```

```python
df=df.loc[df['Distância']!=0]
```

```python
df['Velocidade']=df['Distância']/df['Tempo em movimento']
```
